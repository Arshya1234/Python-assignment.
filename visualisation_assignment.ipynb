5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
data = np.random.randn(100, 5)
corr = np.corrcoef(data, rowvar=False)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f",
            xticklabels=[f'Feature {i}' for i in range(1, 6)],
            yticklabels=[f'Feature {i}' for i in range(1, 6)])
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot of the Function')
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)
bubble_sizes = df['Population'] * 0.1
plt.figure(figsize=(10, 8))
plt.scatter(df['GDP'], df['Population'], s=bubble_sizes, alpha=0.5)
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: GDP vs Population')
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country)

plt.grid(True)
plt.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=600, height=400)
p.line(x, y, line_width=2)
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd
np.random.seed(42)
data = {
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'sizes': np.random.randint(5, 15, 50),
    'colors': np.random.choice(['red', 'green', 'blue'], 50)
}
df = pd.DataFrame(data)

source = ColumnDataSource(df)
p = figure(title="Scatter Plot with Random Data", width=600, height=400)
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
import pandas as pd
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
data = pd.DataFrame({'Fruits': fruits, 'Counts': counts})
source = ColumnDataSource(data)
p = figure(x_range=data['Fruits'], plot_height=350, title="Fruitscount",toolbar_location=None, tools="")
p.vbar(x='Fruits', top='Counts', width=0.9, source=source)
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.yaxis.major_label_orientation = "horizontal"
p.yaxis.axis_label_text_font_size = "12pt"
p.xaxis.axis_label_text_font_size = "12pt"
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
import numpy as np
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
p.title.text_font_size = "16px"
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.data_heatmap = np.random.rand(10, 10)

xx, yy = np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper
import numpy as np
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
p = figure(title="Heatmap")
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())
p.rect(x=xx.flatten(), y=yy.flatten(), width=1, height=1, fill_color={'field': 'value', 'transform': color_mapper},
       line_color=None)
p.xaxis.axis_label = 'X'
p.yaxis.axis_label = 'Y'
p.title.text_font_size = "16px"

show(p)
```
